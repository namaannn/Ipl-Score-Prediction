```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
##x_train = df.drop(labels='total', axis=1)[df['date'].dt.year<=2016]
#x_test = df.drop(labels='total', axis=1)[df['date'].dt.year>=2017]
#y_train = df[df['date'].dt.year<=2016]['total'].values#
#y_test = df[df['date'].dt.year>=2017]['total'].values#
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
from sklearn.ensemble import RandomForestRegressor
rfc=RandomForestRegressor()
rfc.fit(x_train,y_train)
pre=rfc.predict(x_test)
```

```python
mean_squared_error(y_test,pre)
```

```python
sns.distplot(y_test-pre)
```

```python
mean_absolute_error(y_test,pre)
```

```python
from sklearn.linear_model import Lasso
lr=Lasso()
lr.fit(x_train,y_train)
pre=lr.predict(x_test)
```

```python
sns.distplot(y_test-pre)
```

```python
mean_squared_error(y_test,pre)
```

```python
mean_absolute_error(y_test,pre)
```
